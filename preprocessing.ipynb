```python
import csv
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Load stopwords
stop_words = set(stopwords.words('english'))

# Initialize stemmer
stemmer = PorterStemmer()
```

```python
# Function to preprocess text
def preprocess_text(text):
    # Remove non-alphanumeric characters
    text = re.sub(r'[^a-zA-Z0-9\s$]', '', text)    
    # Tokenize text
    tokens = word_tokenize(text)
    
    # Convert text to lowercase
    tokens = [token.lower() for token in tokens]
    
    # Remove stopwords
    tokens = [token for token in tokens if token not in stop_words]
    
    # Stemming
    tokens = [stemmer.stem(token) for token in tokens]
    
    # Join tokens back into a string
    preprocessed_text = ' '.join(tokens)
    
    return preprocessed_text
```

```python
# Function to preprocess CSV file
def preprocess_csv(input_file, output_file):
    with open(input_file, 'r') as csv_file:
        reader = csv.DictReader(csv_file)
        fieldnames = reader.fieldnames

        with open(output_file, 'w', newline='') as output_csv:
            writer = csv.DictWriter(output_csv, fieldnames=fieldnames)
            writer.writeheader()

            for row in reader:
                # Preprocess text fields
                for field in ['Job Title', 'Education Qualification']:
                    row[field] = preprocess_text(row[field])

                # Write preprocessed row to output file
                writer.writerow(row)
```

```python
# Example usage
input_file = 'hr-data.csv'
output_file = 'preprocessed-hr-data.csv'
preprocess_csv(input_file, output_file)
```
